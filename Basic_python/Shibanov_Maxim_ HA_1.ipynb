Part 1

```python
student_marks = {}

subjects = ["Physics", "Chemistry", "Mathematics", "History", "English"]

# Loop through 5 students
for i in range(1, 6):
    while True:
        try:
            name = input(f"Enter the name of student {i}: ")
            if name.replace(" ", "").isalpha():
                break
            else:
                print('Pleas enter a valid name.')
        except ValueError:
            print("Invalid student name. Please enter a string value.")

    student_marks[name] = {} 
    
    # Loop through subjects to get marks
    for subject in subjects:
        while True:
            try:
                marks = float(input(f"Enter {name}'s mark for {subject}: "))
                if 0 <= marks <= 100:
                    student_marks[name][subject] = marks
                    break
                else:
                    print("Marks should be between 0 and 100. Try again.")
            except ValueError:
                print("Invalid mark. Please enter a numeric value.")
```

```python
# Display the collected marks
def display_marks(data):
    print("\nStudent Marks:")
    for student, marks in data.items():
        print(f"\n{student}:")
        for subject, mark in marks.items():
            print(f"  {subject}: {mark}")

display_marks(student_marks)
```

Part 2


```python
class StudentMarks:
    def __init__(self):
        self.student_marks = {}
        self.subjects = ["Physics", "Chemistry", "Mathematics", "History", "English"]
        self.input_marks()
    
    def get_grade(self, marks):
        if marks >= 90:
            return 'A'
        elif marks >= 80:
            return 'B'
        elif marks >= 70:
            return 'C'
        elif marks >= 60:
            return 'D'
        elif marks >= 50:
            return 'E'
        else:
            return 'F'
    
    def input_marks(self):
        for i in range(1, 6):
            name = input(f"Enter the name of student {i}: ").strip()
            while not name.replace(" ", "").isalpha():
                print("Please enter a valid name containing only alphabetic characters.")
                name = input(f"Enter the name of student {i}: ").strip()
            
            self.student_marks[name] = {}
            
            for subject in self.subjects:
                while True:
                    try:
                        marks = float(input(f"Enter marks for {subject}: "))
                        if 0 <= marks <= 100:
                            self.student_marks[name][subject] = marks
                            break
                        else:
                            print("Marks should be between 0 and 100. Try again.")
                    except ValueError:
                        print("Invalid input. Please enter a numeric value.")
    
    def display_raw_marks(self, student_name=None):
        if student_name:
            if student_name in self.student_marks:
                print(f"\nGrades for {student_name}:")
                for subject, mark in self.student_marks[student_name].items():
                    print(f"  {subject}: {mark}")
            else:
                print("Student not found.")
        else:
            print("\nStudent Marks:")
            for student, marks in self.student_marks.items():
                print(f"\n{student}:")
                for subject, mark in marks.items():
                    print(f"  {subject}: {mark}")
    
    def display_marks_with_scale(self, student_name=None):
        if student_name:
            if student_name in self.student_marks:
                print(f"\nGrades for {student_name}:")
                for subject, mark in self.student_marks[student_name].items():
                    print(f"  {subject}: {mark} ({self.get_grade(mark)})")
            else:
                print("Student not found.")
        else:
            print("\nStudent Marks with Grades:")
            for student, marks in self.student_marks.items():
                print(f"\n{student}:")
                for subject, mark in marks.items():
                    print(f"  {subject}: {mark} ({self.get_grade(mark)})")
```

```python
db = StudentMarks()
db.display_raw_marks()
db.display_marks_with_scale()
```
