```python
import matplotlib.pyplot as plt
```

```python


# Data for the first three rows of the last table
years = [2022, 2023, 2024, 2025]
russia_per_capita = [201, 201, 214, 220]
moscow_per_capita = [494, 505, 482, 536]
st_petersburg_per_capita = [597, 577, 560, 584]

# Plotting
plt.figure(figsize=(6, 2))
plt.plot(years, russia_per_capita, marker='o', linestyle='-', label="Russia")
plt.plot(years, moscow_per_capita, marker='s', linestyle='--', label="Moscow")
plt.plot(years, st_petersburg_per_capita, marker='^', linestyle='-.', label="St. Petersburg")

# Labels and title
plt.xlabel("Year")
plt.title("Retail Space per Capita (sq m per 1,000 residents)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Data for newly commissioned retail space
years = [2022, 2023, 2024, 2025]
newly_commissioned_total = [268, 289, 383, 815]
regional_cities = [146, 202, 210, 315]
moscow_area = [106, 51, 173, 361]
st_petersburg_area = [16, 36, 0, 139]

# Plotting
plt.figure(figsize=(5, 3))
plt.plot(years, newly_commissioned_total, marker='o', linestyle='-', label="Total Russia")
plt.plot(years, regional_cities, marker='s', linestyle='--', label="Regional Cities")
plt.plot(years, moscow_area, marker='^', linestyle='-.', label="Moscow Area")
plt.plot(years, st_petersburg_area, marker='d', linestyle=':', label="St. Petersburg")

# Labels and title
plt.xlabel("Year")
plt.title("Newly Commissioned Retail Space (thousand sq m)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
