```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the translated text
companies = ["Wildberries", "Ozon", "Magnit", "X5 Group", "GK Krasny Yar & Slata", 
             "Svetofor", "Fix Price", "GK DNS", "M.Video-Eldorado"]
revenues = [1669, 844, 2280, 2596, 2581, None, 267, 601, None]  # In billion RUB

growth_rates = [98, 88, 26.2, 18.3, 17.6, None, 22, 7.7, -13.5]  # Percent

# Remove None values for a clean plot
valid_indices = [i for i, rev in enumerate(revenues) if rev is not None]

# Filtered Data
companies_filtered = [companies[i] for i in valid_indices]
revenues_filtered = [revenues[i] for i in valid_indices]
growth_filtered = [growth_rates[i] for i in valid_indices]

# Plot Revenue Comparison
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.bar(companies_filtered, revenues_filtered, color='blue', alpha=0.7, label="Revenue (Billion RUB)")
ax1.set_ylabel("Revenue (Billion RUB)")
ax1.set_xlabel("Company")
ax1.set_title("Revenue and Growth of Top Russian Retailers (2022)")
ax1.set_xticklabels(companies_filtered, rotation=45, ha="right")

# Second Y-Axis for Growth Rate
ax2 = ax1.twinx()
ax2.plot(companies_filtered, growth_filtered, color='red', marker='o', linestyle='-', label="Growth Rate (%)")
ax2.set_ylabel("Growth Rate (%)")

# Legends
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Creating a pie chart for market share distribution

# Filtering out companies with missing revenue values
companies_pie = companies_filtered
revenues_pie = revenues_filtered

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(revenues_pie, labels=companies_pie, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Title
plt.title("Market Share of Top Russian Retailers (2022)")

# Show the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the image

# Real Disposable Income Growth (%) and Inflation (%)
years = [2019, 2020, 2021, 2022, 2023, 2024, 2025]
real_income_growth = [1.2, -2.0, 3.3, 4.5, 7.3, 8.4, 4.0]  # Red line
inflation = [4.5, 3.4, 6.7, 13.8, 9.2, 9.4, 11.0]  # Black line
food_inflation = [5.1, 3.9, 8.4, 14.9, 6.0, 9.2, 9.4]  # Dashed line

# Retail Turnover (Trillion RUB)
years_turnover = [2019, 2020, 2021, 2022, 2023, 2024, 2025]
total_turnover = [33.6, 33.9, 39.5, 42.6, 48.2, 55.6, 62.4]
food_turnover = [16.1, 16.3, 18.6, 21.0, 23.0, 26.4, 30.1]
nonfood_turnover = [17.5, 17.6, 20.9, 21.5, 25.1, 29.1, 32.3]

# Plot Real Income Growth vs Inflation
plt.figure(figsize=(8, 5))
plt.plot(years, real_income_growth, marker='o', linestyle='-', color='red', label="Real Disposable Income Growth (%)")
plt.plot(years, inflation, marker='o', linestyle='-', color='black', label="Inflation (%)")
plt.plot(years, food_inflation, marker='o', linestyle='--', color='gray', label="Food Inflation (%)")

plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.title("Growth of Real Disposable Income and Inflation (YoY)")
plt.legend()
plt.grid(True)

plt.show()

# Plot Retail Turnover as a Stacked Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(years_turnover, nonfood_turnover, color='gray', label="Non-Food Goods")
plt.bar(years_turnover, food_turnover, bottom=nonfood_turnover, color='red', label="Food Goods")

plt.xlabel("Year")
plt.ylabel("Retail Turnover (Trillion RUB)")
plt.title("Retail Turnover in Russia (2019-2025)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Plot Retail Turnover as a Stacked Bar Chart with Values
plt.figure(figsize=(8, 5))
bars1 = plt.bar(years_turnover, nonfood_turnover, color='gray', label="Non-Food Goods")
bars2 = plt.bar(years_turnover, food_turnover, bottom=nonfood_turnover, color='red', label="Food Goods")

# Add labels to bars
for bar, value in zip(bars1, nonfood_turnover):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f"{value:.1f}", ha='center', va='center', color='white', fontsize=10)

for bar, value, base in zip(bars2, food_turnover, nonfood_turnover):
    plt.text(bar.get_x() + bar.get_width()/2, base + bar.get_height()/2, f"{value:.1f}", ha='center', va='center', color='white', fontsize=10)

# Labels and title
plt.xlabel("Year")
plt.ylabel("Retail Turnover (Trillion RUB)")
plt.title("Retail Turnover in Russia (2019-2025)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Plot Retail Turnover as a Stacked Bar Chart with Total Values
plt.figure(figsize=(8, 5))
bars1 = plt.bar(years_turnover, nonfood_turnover, color='gray', label="Non-Food Goods")
bars2 = plt.bar(years_turnover, food_turnover, bottom=nonfood_turnover, color='red', label="Food Goods")

# Add labels to individual sections
for bar, value in zip(bars1, nonfood_turnover):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, f"{value:.1f}", ha='center', va='center', color='white', fontsize=10)

for bar, value, base in zip(bars2, food_turnover, nonfood_turnover):
    plt.text(bar.get_x() + bar.get_width()/2, base + bar.get_height()/2, f"{value:.1f}", ha='center', va='center', color='white', fontsize=10)

# Add total values on top of each bar
total_values = [food + nonfood for food, nonfood in zip(food_turnover, nonfood_turnover)]
for bar, total in zip(bars2, total_values):
    plt.text(bar.get_x() + bar.get_width()/2, total + 1, f"{total:.1f}", ha='center', va='bottom', color='black', fontsize=11, fontweight='bold')

# Labels and title
plt.xlabel("Year")
plt.ylabel("Retail Turnover (Trillion RUB)")
plt.title("Retail Turnover in Russia (2019-2025)**")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
